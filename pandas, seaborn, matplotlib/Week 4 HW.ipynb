```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# A) create scatterplot with workers tips CSV file
tips = pd.read_csv('workerstips.csv')
sns.scatterplot(x='total_bill',y='tip',data=tips)
```

```python
# B) create scatterplot with workers tips CSV file and make variations
tips = pd.read_csv('workerstips.csv')
sns.scatterplot(x='total_bill',y='tip',hue='smoker',style='day',
                sizes=(10,300),data=tips)
```

```python
# C) Create barplot to show average tips per day of the week
tips = pd.read_csv('workerstips.csv')
sns.barplot(y='tip',x='day',data=tips)
```

```python
# D) split the distribution by lunch and dinner
tips = pd.read_csv('workerstips.csv')
sns.scatterplot(x='total_bill',y='tip',hue='time',style='day',
                sizes=(10,300),data=tips)
```

```python
# E) use flightdata.csv to create a lineplot with avg num passengers
# per year per month
flights = pd.read_csv('flightsData.csv')
sns.lineplot(data=flights,x='year',y='passengers',hue='month',style='month'
            ,sizes=(10,300))
```

```python
#find the total passengers per year and plot it
flights = pd.read_csv('flightsData.csv')
pass_year = flights.groupby('year').sum()
print('Here are the years with their total passengers:\n\n',pass_year)
sns.lineplot(data=pass_year, x='year',y='passengers')
```
