```python
# Bubble sort algorithm for various sized lists
data1 = []
for i in range(0,10000):
    x = random.randint(0,10000)
    data1.append(x)
size1a = len(data1)

data1a = []
for i in range(0,30000):
    x = random.randint(0,30000)
    data1a.append(x)

data1b = []
for i in range(0,50000):
    x = random.randint(0,50000)
    data1b.append(x)

data1c = []
for i in range(0,70000):
    x = random.randint(0,70000)
    data1c.append(x)

data1d = []
for i in range(0,90000):
    x = random.randint(0,90000)
    data1d.append(x)

#bubbleSort(data1)
#print(data1)

# See how long it takes to sort the data
#Start the clock
start_time = time.time()
#Run bubble sort
bubbleSort(data1)
#stop clock and calculate the time it took
stop_time = time.time() - start_time
print("the bubblesort took %s seconds" % stop_time)

#Start the clock
start_time1 = time.time()
#Run bubble sort
bubbleSort(data1a)
#stop clock and calculate the time it took
stop_time1 = time.time() - start_time1
print("the bubblesort took %s seconds" % stop_time1)

#Start the clock
start_time2 = time.time()
#Run bubble sort
bubbleSort(data1b)
#stop clock and calculate the time it took
stop_time2 = time.time() - start_time2
print("the bubblesort took %s seconds" % stop_time2)

#Start the clock
start_time3 = time.time()
#Run bubble sort
bubbleSort(data1c)
#stop clock and calculate the time it took
stop_time3 = time.time() - start_time3
print("the bubblesort took %s seconds" % stop_time3)

#Start the clock
start_time4 = time.time()
#Run bubble sort
bubbleSort(data1d)
#stop clock and calculate the time it took
stop_time4 = time.time() - start_time4
print("the bubblesort took %s seconds" % stop_time4)

x = [10000, 30000, 50000, 70000, 90000]
y = [stop_time, stop_time1, stop_time2, stop_time3, stop_time4]

plt.plot(x,y)
plt.xlabel('Range')
plt.ylabel('Time (in seconds)')
plt.title('Bubblesort time')
plt.show()
```
